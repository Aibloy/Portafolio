### Correlaciones 

```python
#CORRELATION
def top_correlation (df,n):
    corr_matrix = df.corr(method="spearman")
    correlation = (corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))
                 .stack()
                 .sort_values(ascending=False))
    correlation = pd.DataFrame(correlation).reset_index()
    correlation.columns=["Variable_1","Variable_2","Correlacion"]
    correlation = correlation.reindex(correlation.Correlacion.abs().sort_values(ascending=False).index).reset_index().drop(["index"],axis=1)
    return correlation.head(n)
    
display(top_correlation(MUSIC,3))

display(top_correlation(MOVIE,3))

display(top_correlation(HOBBY,3))

display(top_correlation(FOBIA,3))

display(top_correlation(PERSONALITY,3))

display(top_correlation(SPEND,3))
```

# PREDICT HAPPINES 

```python
#Encoding ordinal data

df = pd.read_csv("D:/DATASET/youngpeople.csv")

code = {"Punctuality":
        {
        "i am often running late":0,  
        'i am often early': 1, 
        'i am always on time':2},
        
        "Lying":
        {"never":0,
        "only to avoid hurting someone":1,
        "sometimes":2,
        "everytime it suits me":3
       },
        "Internet usage":
        {
        "no time at all":0,  
        'less than an hour a day': 1, 
        'few hours a day':2,
        "most of the day":3
        },
        "Smoking":
        {'never smoked':0,  
        'tried smoking': 1, 
        'former smoker':2,
        "current smoker":3,
        "never":0,
        "social drinker":1,
        "drink a lot":2
        },
        "Alcohol":
        {"never":0,
         "social drinker":1,
         "drink a lot":2
        },
        "Education":
        {"currently a primary school pupil":0,
         "primary school":1,
         "secondary school":2,
         "college/bachelor degree":3,
         "masters degree":4,
         "doctorate degree":5
       }}

df.replace(code, inplace=True)
```

```python
#Pre procesado datos 
#Eliminar valores nulos


#Delete extreme value
condition=df["Height"]<100
df.drop(df[condition]. index, axis=0, inplace=True)


#Borrar valores nulos 
df = df.dropna()




#.................................................................................................................
#Codificar datos categÃ³ricos
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer

df = df.apply(LabelEncoder().fit_transform)

#Seleccionar variables.....................................................................................................
Tarjet = "Happiness in life"

X = df.drop(Tarjet,axis=1).values
Y = df[Tarjet].values

Xn = X.shape[1]             #Numero de valores de entrada de X 
Prueba = df.drop(Tarjet,axis=1).head(0)       #Valores de entrada 


# Dividir el data set en conjunto de entrenamiento y conjunto de testing............................................
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 0)

# Escalado de variables....................................................................................................
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

```python
Xn
```

```python
#PLANTILLA: ANN - REGRESIÃ“N 
from keras.models import Sequential
from keras.layers import Dense


ANN = Sequential()
ANN.add(Dense(input_dim=Xn, 
                  units = 100,
                  activation='relu'))

ANN.add(Dense(units = 100, activation = 'relu'))
    
ANN.add(Dense(units = 1))
    
ANN.compile(optimizer = 'adam', loss = 'mean_squared_error')

ANN.fit(X_train, Y_train,  batch_size = 50, epochs = 100, verbose=0)

Y_pred = ANN.predict(X_test)

plt.plot(Y_test, color = 'red', label = 'Real')
plt.plot(Y_pred, color = 'blue', label = 'PredicciÃ³n', alpha=0.55)
plt.legend()
plt.show()
```
